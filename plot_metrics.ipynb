```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
plt.rcParams.update({'font.size': 16})
```

```python
subjects = [i for i in range(0, 10)]
subjects.append(11)

s_names = [f"S{i}" for i in range(1,11)]
s_names.append("Cross-Sub")

conditions = ["Inner", "Pronounced", "Visualized"]
```

```python
accuracy_scores = []
for subject in subjects:

    with open(f"classifier_{subject}.pickle", "rb") as f:
        clf = pickle.load(f)
        
    accuracy_scores.append(clf['validation_data_classification_report']["accuracy"]*100) 
```

```python
df_accuracy = pd.DataFrame({"Subject": s_names, "Accuracy": accuracy_scores})

plt.figure(figsize=(15, 8))
plt.title("Accuracy scores")
sns.barplot(x = 'Subject', y = 'Accuracy', data = df_accuracy)
plt.grid()
plt.savefig('Accuracy.png', bbox_inches='tight')
plt.show()
```

```python
list_s = []
list_c = []
list_p = []

for i, subject in enumerate(subjects):

    with open(f"classifier_{subject}.pickle", "rb") as f:
        clf = pickle.load(f)
        
    for _ in range(3):
        list_s.append(s_names[i])
        
    for j, c in enumerate(conditions):
        list_c.append(c)
        list_p.append(clf['validation_data_classification_report'][c]['precision']*100) 

df_precision = pd.DataFrame({"Subject": list_s, "Condition": list_c, "Precision": list_p})

print(df_precision.shape)
```

```python
plt.figure(figsize=(15, 10))
plt.title("Precision scores")
sns.barplot(x = "Subject", y = "Precision", hue = "Condition", data = df_precision)
plt.grid()
plt.savefig('Precision.png', bbox_inches='tight')
plt.show()
```

```python
list_s2 = []
list_c2 = []
list_r = []

for i, subject in enumerate(subjects):

    with open(f"classifier_{subject}.pickle", "rb") as f:
        clf = pickle.load(f)
        
    for _ in range(3):
        list_s2.append(s_names[i])
        
    for j, c in enumerate(conditions):
        list_c2.append(c)
        list_r.append(clf['validation_data_classification_report'][c]['recall']*100) 

df_recall = pd.DataFrame({"Subject": list_s2, "Condition": list_c2, "Recall": list_r})

print(df_recall.shape)
```

```python
plt.figure(figsize=(15, 10))
plt.title("Recall scores")
sns.barplot(x = "Subject", y = "Recall", hue = "Condition", data = df_recall)
plt.grid()
plt.savefig('Recall.png', bbox_inches='tight')
plt.show()
```

```python
list_s3 = []
list_c3 = []
list_f = []

for i, subject in enumerate(subjects):

    with open(f"classifier_{subject}.pickle", "rb") as f:
        clf = pickle.load(f)
        
    for _ in range(3):
        list_s3.append(s_names[i])
        
    for j, c in enumerate(conditions):
        list_c3.append(c)
        list_f.append(clf['validation_data_classification_report'][c]['f1-score']*100) 

df_f1 = pd.DataFrame({"Subject": list_s3, "Condition": list_c3, "F1-score": list_f})

print(df_f1.shape)
```

```python
plt.figure(figsize=(15, 10))
plt.title("F1 scores")
sns.barplot(x = "Subject", y = "F1-score", hue = "Condition", data = df_f1)
plt.grid()
plt.savefig('f1.png', bbox_inches='tight')
plt.show()
```
